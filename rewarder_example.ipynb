```python
from rewarder import Rewarder
import json
import pandas as pd
from tqdm import tqdm
```

```python
## example df
df = pd.read_csv('./data/ap/expanded_and_labeled_queries_with_pubmed_ap_n100_200candidates.csv')
df = df.sample(10)
```

```python
def calculate_reward(rewarder: Rewarder, base_query_column='Q_name', expanded_query_colum='new_q'):
    def eval(row):
        base_query = row[base_query_column]        
        new_query = row[expanded_query_colum]
        return rewarder.reward_one(new_query, base_query)
    return eval
```

```python
metric = 'recall'
at_n = 100
rewarder = Rewarder(
        "./data/cord_qrels.json", n_results=at_n, metric=metric
    )

tqdm.pandas()
df['new_query_recall'] = df.progress_apply(calculate_reward(rewarder), axis=1)
df
```
